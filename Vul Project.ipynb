#Pre-trained

```python
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModelForMaskedLM, DataCollatorForLanguageModeling, create_optimizer
from transformers import pipeline

model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = TFAutoModelForMaskedLM.from_pretrained(model_checkpoint)

def tokenize_function(examples):
    result = tokenizer(examples["description"])
    if tokenizer.is_fast:
        result["word_ids"] = [result.word_ids(i) for i in range(len(result["input_ids"]))]
    return result

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["description"])

chunk_size=128

def group_texts(examples):
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    total_length = (total_length // chunk_size) * chunk_size
    result = {
        k: [t[i : i + chunk_size] for i in range(0, total_length, chunk_size)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
tokenized_dataset
```

```python
tokenized_dataset = tokenized_dataset.map(group_texts,batched=True)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=True, mlm_probability=0.15)

train_size = 10_000
test_size = int(0.1 * train_size)

downsampled_dataset = tokenized_dataset.train_test_split(
    train_size=train_size, test_size=test_size, seed=42
)
downsampled_dataset
```

```python
tf_train_dataset = model.prepare_tf_dataset(
    downsampled_dataset["train"],
    collate_fn=data_collator,
    shuffle=True,
    batch_size=32,
)

tf_eval_dataset = model.prepare_tf_dataset(
    downsampled_dataset["test"],
    collate_fn=data_collator,
    shuffle=False,
    batch_size=32,
)
```

```python
tf_train_dataset
```

```python
tf_eval_dataset
```

##perplexity

```python
import math

tf.keras.mixed_precision.set_global_policy("mixed_float16")

num_train_steps = len(tf_train_dataset)
optimizer, schedule = create_optimizer(
    init_lr=2e-5,
    num_warmup_steps=1_000,
    num_train_steps=num_train_steps
)

model.compile(optimizer=optimizer)

perplexity={}
eval_loss = model.evaluate(tf_eval_dataset)
perplexity['pre-trained']=math.exp(eval_loss)
print(f"Perplexity: {math.exp(eval_loss):.2f}")

model.fit(tf_train_dataset, validation_data=tf_eval_dataset)

eval_loss = model.evaluate(tf_eval_dataset)
perplexity['fine-tuned']=math.exp(eval_loss)
print(f"Perplexity: {math.exp(eval_loss):.2f}")

model.save_pretrained('cyber_model')
tokenizer.save_pretrained('cyber_tokenizer')

fill_mask = pipeline('fill-mask', model='cyber_model', tokenizer='cyber_tokenizer')
```

##test sentences

```python
test_sentences = [
    "It may be used to look for patterns associated with [MASK] content designed to collect host information from visitors."]

print("Fine-tuned model predictions:")
for sentence in test_sentences:
    predictions = fill_mask(sentence)
    for prediction in predictions:
        print(f">>> {prediction['sequence']}")

pretrained_fill_mask = pipeline('fill-mask', model=model_checkpoint, tokenizer=model_checkpoint)
print("Pre-trained model predictions:")
for sentence in test_sentences:
    predictions = pretrained_fill_mask(sentence)
    for prediction in predictions:
        print(f">>> {prediction['sequence']}")
```

##Sequence Classification

```python
import pandas as pd
import re

file_path = '/content/drive/MyDrive/babe/cve_data.csv'
data = pd.read_csv(file_path)

data_cleaned = data.dropna(subset=['description', 'cwe_id']).copy()

data_cleaned.loc[:, 'cwe_id'] = data_cleaned['cwe_id'].astype(str)

data_cleaned = data_cleaned[data_cleaned['cwe_id'].apply(lambda x: bool(re.search(r'\d', x)))]

def trim_cwe_id(cwe_id):
    if '-' in cwe_id:
        return cwe_id.split('-')[1]
    return cwe_id

data_cleaned.loc[:, 'cwe_id_trimmed'] = data_cleaned['cwe_id'].apply(trim_cwe_id)

print(data_cleaned['cwe_id_trimmed'].unique())

data_cleaned.loc[:, 'cwe_id_trimmed'] = pd.to_numeric(data_cleaned['cwe_id_trimmed'], errors='coerce')

data_cleaned = data_cleaned.dropna(subset=['cwe_id_trimmed'])

are_all_numeric = data_cleaned['cwe_id_trimmed'].apply(lambda x: isinstance(x, (int, float))).all()
print(f"All values in 'cwe_id_trimmed' are numeric: {are_all_numeric}")

data_cleaned.loc[:, 'cwe_id_trimmed'] = data_cleaned['cwe_id_trimmed'].astype(int)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
import tensorflow as tf

data = df_cleaned.sample(n=10000, random_state=42)

unique_labels = data['cwe_id_trimmed'].unique()
label_mapping = {label: idx for idx, label in enumerate(unique_labels)}
data['cwe_id_trimmed'] = data['cwe_id_trimmed'].map(label_mapping)

num_labels = len(unique_labels)
print(f"Number of unique labels (num_labels): {num_labels}")
train_texts, val_texts, train_labels, val_labels = train_test_split(data['description'], data['cwe_id_trimmed'], test_size=0.2, random_state=42)

tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

train_encodings = tokenizer(train_texts.tolist(), truncation=True, padding=True)
val_encodings = tokenizer(val_texts.tolist(), truncation=True, padding=True)

train_dataset = tf.data.Dataset.from_tensor_slices((dict(train_encodings), train_labels)).shuffle(10000).batch(32)
val_dataset = tf.data.Dataset.from_tensor_slices((dict(val_encodings), val_labels)).batch(32)
```

```python
from transformers import TFDistilBertForSequenceClassification, create_optimizer

tf.keras.mixed_precision.set_global_policy("float32")

num_train_steps = len(train_dataset)*3
optimizer, lr_schedule = create_optimizer(init_lr=5e-3, num_warmup_steps=0, num_train_steps=num_train_steps)

distilbert_model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=num_labels)

distilbert_model.compile(optimizer=optimizer, loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

distilbert_model.fit(train_dataset, validation_data=val_dataset, epochs=3)

distilbert_eval_results = distilbert_model.evaluate(val_dataset)
print("DistilBERT Evaluation Results:", distilbert_eval_results)
```

```python
cyber_model_path = 'cyber_model'

cyber_model = TFDistilBertForSequenceClassification.from_pretrained(cyber_model_path, num_labels=num_labels)

cyber_model.compile(optimizer=optimizer, loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))

cyber_optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)

cyber_model.compile(optimizer=cyber_optimizer, loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

cyber_model.fit(train_dataset, validation_data=val_dataset, epochs=3)

cyber_model_eval_results = cyber_model.evaluate(val_dataset)
print("Cyber Model Evaluation Results:", cyber_model_eval_results)
```
